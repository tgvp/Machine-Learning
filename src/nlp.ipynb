# Natural Language Processing - NLP

This notebook showcases the use of NLP techniques to analyze text data.

## Word count
- First, we will count the number of words in the literary work of Machado de Assis, a famous Brazilian writer. The text is in Portuguese.

- The words will be counted without the use of any library, only with the use of Python.

```python
with open('../data/casmurro.txt', 'r', encoding='utf-8') as f:
    casmurro = f.read().splitlines()
    
casmurro
```

```python
word_count = {}

# strip punctuation and convert to lowercase
for line in casmurro:
    for word in line.split():
        word = word.strip('.,!?:;--«»"\'').lower()
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
            
word_count
```

```python
print(f'Number of words in the dictionary: {len(word_count)}')
print(f'Most frequent word: {max(word_count, key=word_count.get)} with {max(word_count.values())} occurrences')
print(f'Percentage of the most frequent word: {max(word_count.values()) / sum(word_count.values()) * 100:.2f}%')
print(f'Top ten words: {sorted(word_count, key=word_count.get, reverse=True)[:10]}')
print(f'Percentage of the sum of top ten words: {sum(list(sorted(word_count.values(), reverse=True))[:10]) / sum(word_count.values()) * 100:.2f}%')
```

```python
import matplotlib.pyplot as plt

plt.bar(word_count.keys(), word_count.values())
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = px.histogram(x=list(word_count.keys()), y=list(word_count.values()))
fig.update_layout(title='Word Count Histogram', xaxis_title='Word', yaxis_title='Count')
fig.show()
```

### Results:

After counting the words in **Dom Casmurro**, we can see that **9591 unique words** were counted with `que` being the **most frequent word**, which corresponds to **4.08%** of the words alone and the percentage of the **top ten words** is **23.41%**.

The **top ten words** are: `['que', 'a', 'e', 'de', 'o', 'não', 'um', 'é', 'os', 'da']` which as we can see does not provide much meaning to the text.
