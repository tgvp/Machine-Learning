### Normalization

```python
import pandas as pd
from sklearn.datasets import make_classification
```

```python
# criando dataframe com 1000 linhas e 10 colunas
dados = sklearn.datasets.make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=5, n_classes=2, n_clusters_per_class=2, random_state=0)

df = pd.DataFrame(dados[0])
df
```

```python
def normalizacao(x):
    return (x - x.mean()) / (.00001 + x.std()) # evitando divisão por zero

df_normalized = df.apply(normalizacao)
df_normalized
```
